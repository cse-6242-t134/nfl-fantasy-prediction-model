```python
df_filtered = df[df["predictions"] >= 10]

df_filtered["abs_pred_err"] = abs(df_filtered["fantasy_points"] - df_filtered["predictions"])


df_filtered["abs_pred_err_pct"] = df_filtered["abs_pred_err"] / df_filtered["predictions"]
df_filtered["abs_pred_err_pct"].mean()
```
